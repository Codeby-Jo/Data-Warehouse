```python
import numpy as np
```

```python
products = ["Laptop", "Mobile", "Tablet"]
regions = ["Chennai", "Delhi", "Mumbai"]
years = [2023, 2024]
```

```python
sales = np.array([
    [[120,150],[90,110],[70,95]],
    [[200,240],[180,210],[150,190]],
    [[80,100],[60,75],[50,65]]
])
```

```python
def slice_cube(cube, axis, index):
    return np.take(cube, index, axis=axis)
```

```python
print(slice_cube(sales, axis=1, index=0))
```

```python
def dice_cube(cube, p, r, y):
    return cube[np.ix_(p, r, y)]
```

```python
print(dice_cube(sales, [0,1], [0,2], [1]))
```

```python
rollup = np.sum(sales, axis=2)
print(rollup)
```

```python
drill = sales[:, :, 1]
print(drill)
```

```python
class OLAPCube:
    def __init__(self, cube):
        self.cube = cube

    def slice(self, axis, index):
        return np.take(self.cube, index, axis=axis)

    def dice(self, p, r, y):
        return self.cube[np.ix_(p, r, y)]

    def rollup(self, axis):
        return np.sum(self.cube, axis=axis)

    def drilldown(self, axis, index):
        return np.take(self.cube, index, axis=axis)

cube = OLAPCube(sales)
```

```python
print("========== SLICE OPERATION ==========")
print(cube.slice(axis=1, index=0))

print("\n========== DICE OPERATION ==========")
print(cube.dice([0,1], [0,2], [1]))

print("\n========== ROLL-UP OPERATION ==========")
print(cube.rollup(axis=2))

print("\n========== DRILL-DOWN OPERATION ==========")
print(cube.drilldown(axis=2, index=1))
```
